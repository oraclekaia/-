#### Define input transform

```python
mean = [0.45, 0.45, 0.45]
std = [0.225, 0.225, 0.225]
frames_per_second = 30
model_transform_params  = {
    "x3d_xs": {
        "side_size": 182,
        "crop_size": 182,
        "num_frames": 4,
        "sampling_rate": 12,
    },
    "x3d_s": {
        "side_size": 182,
        "crop_size": 182,
        "num_frames": 13,
        "sampling_rate": 6,
    },
    "x3d_m": {
        "side_size": 256,
        "crop_size": 256,
        "num_frames": 16,
        "sampling_rate": 5,
    }
}

# Get transform parameters based on model
transform_params = model_transform_params[model_name]

# Note that this transform is specific to the slow_R50 model.
transform =  ApplyTransformToKey(
    key="video",
    transform=Compose(
        [
            UniformTemporalSubsample(transform_params["num_frames"]),
            Lambda(lambda x: x/255.0),
            NormalizeVideo(mean, std),
            ShortSideScale(size=transform_params["side_size"]),
            CenterCropVideo(
                crop_size=(transform_params["crop_size"], transform_params["crop_size"])
            )
        ]
    ),
)

# The duration of the input clip is also specific to the model.
clip_duration = (transform_params["num_frames"] * transform_params["sampling_rate"])/frames_per_second
```

Load the video and transform it to the input format required by the model.

```python
import csv
import os

import pandas as pd
from PIL import Image

import torch
from torch.utils.data import Dataset, DataLoader

class X3DDataset(Dataset):
    def __init__(self, csv_file, transform):
        self.video_paths = []
        self.labels = []
        self.transform = transform

        with open(f'{csv_file}', newline='') as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
              # print(row['image/gif'])
              if(row['image/gif'] == '1'):
                self.video_paths.append(f"{config.mp4_path}/{row['tokenId']}.mp4")
                # if mode != 'test':
#                 self.labels.append(float(row['avg_eth_norm_std']))
                self.labels.append(float(row['last_eth_norm_std']))
                # self.labels.append(row['max_eth_norm_std'])
                # self.labels.append(row['last_eth_norm_std'])
    def __len__(self):
        return len(self.video_paths)
    
    def __getitem__(self, index):
        video_path = self.video_paths[index]
        label = self.labels[index]

        # The duration of the input clip is also specific to the model.
        clip_duration = (transform_params["num_frames"] * transform_params["sampling_rate"])/frames_per_second

        # Select the duration of the clip to load by specifying the start and end duration
        # The start_sec should correspond to where the action occurs in the video
        start_sec = 0
        end_sec = start_sec + clip_duration
        # Initialize an EncodedVideo helper class and load the video
        video = EncodedVideo.from_path(video_path)

        # Load the desired clip
        video_data = video.get_clip(start_sec=start_sec, end_sec=end_sec)

        # Apply a transform to normalize the video input
        video_data = self.transform(video_data)
        return video_data["video"], label
```

```python
import torch.nn as nn
# from transformers import AdamW
import torch.optim as optim
from torch.optim import AdamW

# 1e-3, 5e-4, 1e-4, 2e-5, 3e-6, 3e-7
lr = 2e-5
# 0.05
weight_decay = 0.04

optimizer = AdamW(model.parameters(), lr = lr, betas=(0.9, 0.99), weight_decay=weight_decay)
# criterion = nn.CrossEntropyLoss()
criterion = nn.MSELoss(reduction='sum')
# criterion = nn.MSELoss()
criterion = criterion.to(device)
```

```python
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import math
import pandas as pd

train_losses_list = []
val_losses_list = []

def train(model, data, optimizer, criterion):
    model.train()

    pred_list = torch.tensor([])
    label_list = torch.tensor([])
    # pred_list = []
    # label_list = []
    epoch_losses = 0
    epoch_acc = 0
    total = 0
    i = 0

    for mp4, labels in tqdm(data, total=len(data)):
        # mp4 = [i.to(device)[None, ...] for i in mp4]
        # mp4 = [i.to(device) for i in mp4]
        mp4 = mp4.to(device)
        labels = labels.float().to(device)
        
        optimizer.zero_grad()

        outputs = model(mp4)

        labels = labels.view(-1, 1)
        loss = criterion(outputs, labels)
        
        loss.backward()
        optimizer.step()
        
        epoch_losses += loss.item()
        train_losses_list.append(loss.item())
          # epoch_acc += acc.item()
        total += labels.size(0)

        # print(outputs.shape)      
        y_bar = outputs.squeeze().cpu()
        # print(y_bar.shape)
        # print(pred_list.shape)
        y = labels.squeeze().cpu()
        # print(y.shape)

        pred_list = torch.cat((pred_list, y_bar), 0)
        label_list = torch.cat((label_list, y), 0)
        # pred_list.append(y_bar)
        # label_list.append(y)
    # print(pred_list)
    # print(label_list)
    mse = mean_squared_error(label_list.detach().numpy(), pred_list.detach().numpy())
    r2 = r2_score(label_list.detach().numpy(), pred_list.detach().numpy())
    # mse = mean_squared_error(label_list, pred_list)
    # r2 = r2_score(label_list, pred_list)

    return epoch_losses / total, mse, r2

def val(model, data, criterion, log_loss=False):
    model.eval()

    pred_list = torch.tensor([])
    label_list = torch.tensor([])
    # pred_list = []
    # label_list = []
    epoch_losses = 0
    epoch_acc = 0
    total = 0
    i = 0
    with torch.no_grad():
        for mp4, labels in tqdm(data, total=len(data)):
            # mp4 = [i.to(device)[None, ...] for i in mp4]
            # mp4 = [i.to(device) for i in mp4]
            mp4 = mp4.to(device)
            labels = labels.float().to(device)

            outputs = model(mp4)

            labels = labels.view(-1, 1)
            loss = criterion(outputs, labels)

            epoch_losses += loss.item()
            if log_loss:
                val_losses_list.append(loss.item())
            # epoch_acc += acc.item()
            total += labels.size(0)
            
            # print(outputs.shape)
            y_bar = outputs.squeeze().cpu()
            # print(y_bar.shape)
            y = labels.squeeze().cpu()
            # print(y.shape)
            pred_list = torch.cat((pred_list, y_bar), 0)
            label_list = torch.cat((label_list, y), 0)
            # pred_list.append(y_bar)
            # label_list.append(y)
            
    mse = mean_squared_error(label_list.detach().numpy(), pred_list.detach().numpy())
    r2 = r2_score(label_list.detach().numpy(), pred_list.detach().numpy())
    # mse = mean_squared_error(label_list, pred_list)
    # r2 = r2_score(label_list, pred_list)

    return epoch_losses / total, mse, r2

class Meter:
    def __init__(self):
        self.train_loss_list = []
        # self.train_acc_list = []
        self.val_loss_list = []
        # self.val_acc_list = []
        self.train_r2_list = []
        self.val_r2_list = []
        # self.train_mse_list = []
        # self.val_mse_list = []

    # def update(self, train_loss, train_acc, val_loss, val_acc):
    def update(self, train_loss, val_loss, train_r2, val_r2):
        self.train_loss_list.append(train_loss)
        # self.train_acc_list.append(train_acc)
        self.val_loss_list.append(val_loss)
        # self.val_acc_list.append(val_acc)
        self.train_r2_list.append(train_r2)
        self.val_r2_list.append(val_r2)

        # self.train_mse_list.append(train_mse)
        # self.val_mse_list.append(val_mse)

    def plot(self, max_epoch):
        x = range(len(self.train_loss_list))
        plt.plot(x, self.train_loss_list)
        plt.plot(x, self.val_loss_list, color='r')

        plt.legend(['train_loss', 'val_loss'])
        plt.savefig('./data/X3D_s_mp4_last_MSELoss_230606_lr_{:06}_wd_{:02}_epoch_{:03}.png'.format(lr, weight_decay, max_epoch), transparent=True)
        plt.show()

        # plt.plot(x, self.train_acc_list)
        # plt.plot(x, self.val_acc_list, color='r')
        plt.plot(x, self.train_r2_list)
        plt.plot(x, self.val_r2_list, color='r')
        plt.legend(['train_r2', 'val_r2'])
        plt.savefig('./data/X3D_s_mp4_last_R2(MSE)_230606_lr_{:06}_wd_{:02}_epoch_{:03}.png'.format(lr, weight_decay, max_epoch), transparent=True)
        plt.show()

        # plt.plot(x, self.train_mse_list)
        # plt.plot(x, self.val_mse_list, color='r')
        # plt.legend(['train_mse', 'val_mse'])
        # plt.savefig('./data/VGG16_NFT_MSE_230224_lr_{:06}_wd_{:02}.png'.format(lr, weight_decay), transparent=True)
        # plt.show()
```

```python
#########################################################################
#                          Hyper-parameters                             #
#########################################################################
max_epoch = 50
log_interval = 1

# for early stopping
best_loss = float('inf')
count = 0
best_acc = 0
best_r2 = 0
best_mse = 1000
#########################################################################
#                          End of your code                             #
#########################################################################

m = Meter()

for epoch in range(1, max_epoch + 1):
    # train_loss, train_acc = train(model, NFT_train_data, optimizer, criterion)
    # val_loss, val_acc = val(model, NFT_val_data, criterion, log_loss=True)
    train_loss, train_mse, train_r2 = train(model, mp4_train_data, optimizer, criterion)
    val_loss, val_mse, val_r2 = val(model, mp4_val_data, criterion, log_loss=True)


    if epoch % log_interval == 0:

        print('Epoch {} train_loss: {} train_MSE: {} train_var: {} train_r2: {}'.format(
            epoch, train_loss, train_mse, train_var, train_r2
        ))
        print('Epoch {} val_loss: {} val_MSE: {} val_var: {} val_r2: {}'.format(
            epoch, val_loss, val_mse, val_var, val_r2
        ))

    # m.update(train_loss, train_acc, val_loss, val_acc)
    m.update(train_loss, val_loss, train_r2, val_r2)

    # model checkpoint

    if (val_mse < val_var) & (train_mse < train_var):
      if (val_mse < best_mse):

          torch.save(model.state_dict(), './ckpts/x3d_s_mp4_last_e{:03}_RMSE_{:.04}_MSE_{:.04}_R2_{:.04}_lr_{:06}_wd_{:02}.pt'.format(epoch, val_loss, val_mse, val_r2, lr, weight_decay))
          best_model = model
          best_mse = val_mse
          best_var = val_var
          best_epoch = epoch
          best_r2 = val_r2
          # best_loss = val_loss
          # best_acc = val_acc
          # min_mse = val_loss
          print('-'*10, 'e', epoch, 'save best model', '-'*10)
```

SlowFast model architectures are based on [1] with pretrained weights using the 8x8 setting
on the Kinetics dataset. 

| arch | depth | frame length x sample rate | top 1 | top 5 | Flops (G) | Params (M) |

| SlowFast | R101  | 8x8                        | 77.90 | 93.27 | 127.20    | 62.83      |
